```python
import nltk
from nltk.util import ngrams
from collections import defaultdict, Counter
```

```python
# 1. Setup and Sample Corpus
nltk.download('punkt')
nltk.download('punkt_tab')

corpus = """
Natural language processing is a subfield of linguistics, computer science, and artificial intelligence.
Natural language processing helps computers communicate with humans in their own language.
Artificial intelligence and machine learning are revolutionizing the world of technology.
The world of technology is changing fast with natural language models.
"""
```

```python
def build_trigram_model(text):
    # Preprocessing
    tokens = nltk.word_tokenize(text.lower())

    # Create Tri-grams
    # A tri-gram model looks at pairs of words to predict the third
    trigrams = list(ngrams(tokens, 3))

    # Store frequencies in a nested dictionary: { (word1, word2): {word3: count} }
    model = defaultdict(Counter)

    for w1, w2, w3 in trigrams:
        model[(w1, w2)][w3] += 1

    return model
```

```python
def get_autocomplete_suggestions(model, input_text, num_suggestions=3):
    # Process input to get the last two words
    input_tokens = nltk.word_tokenize(input_text.lower())

    if len(input_tokens) < 2:
        return "Please enter at least two words for tri-gram prediction."

    # Get the last two words as the key
    context = tuple(input_tokens[-2:])

    # Find possible next words
    predictions = model[context]

    # Return the most common predictions
    if not predictions:
        return []

    return [word for word, count in predictions.most_common(num_suggestions)]
```

```python
# 2. Execution
ngram_model = build_trigram_model(corpus)

# Test cases
test_phrases = [
    "natural language",
    "artificial intelligence",
    "the world"
]

print("--- N-Gram Auto-Complete Suggestions ---")
for phrase in test_phrases:
    suggestions = get_autocomplete_suggestions(ngram_model, phrase)
    print(f"Input: '{phrase}' -> Suggestions: {suggestions}")
```
